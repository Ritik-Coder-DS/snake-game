```python
import pygame, sys, random
CELL, COLS, ROWS = 20, 32, 24
W, H = COLS*CELL, ROWS*CELL
pygame.init()
s = pygame.display.set_mode((W,H))
clk = pygame.time.Clock()
font = pygame.font.SysFont(None, 20)

def rnd(snake):
    while True:
        p = (random.randrange(COLS), random.randrange(ROWS))
        if p not in snake: return p

def draw(texts, snake, food, score):
    s.fill((10,10,30))
    for x in range(0,W,CELL): pygame.draw.line(s,(25,25,50),(x,0),(x,H))
    for y in range(0,H,CELL): pygame.draw.line(s,(25,25,50),(0,y),(W,y))
    fx,fy = food
    pygame.draw.rect(s,(200,50,50),(fx*CELL+2,fy*CELL+2,CELL-4,CELL-4),border_radius=4)
    for i,(x,y) in enumerate(snake):
        rect = (x*CELL+1,y*CELL+1,CELL-2,CELL-2)
        col = (0,230,0) if i else (0,200,0)
        pygame.draw.rect(s,col,rect,border_radius=4)
    y=8
    for t in texts:
        s.blit(font.render(t,True,(230,230,230)),(8,y)); y+=18
    s.blit(font.render(f"Score: {score}",True,(230,230,230)),(W-110,8))
    pygame.display.flip()

def game():
    dir = (1,0)
    snake = [(COLS//2, ROWS//2),(COLS//2-1, ROWS//2),(COLS//2-2, ROWS//2)]
    food = rnd(snake)
    score = 0
    speed = 10
    running = True
    while running:
        clk.tick(speed)
        for e in pygame.event.get():
            if e.type==pygame.QUIT: pygame.quit(); sys.exit()
            if e.type==pygame.KEYDOWN:
                if e.key in (pygame.K_UP, pygame.K_w) and dir!=(0,1): dir=(0,-1)
                if e.key in (pygame.K_DOWN, pygame.K_s) and dir!=(0,-1): dir=(0,1)
                if e.key in (pygame.K_LEFT, pygame.K_a) and dir!=(1,0): dir=(-1,0)
                if e.key in (pygame.K_RIGHT, pygame.K_d) and dir!=(-1,0): dir=(1,0)
                if e.key==pygame.K_ESCAPE: pygame.quit(); sys.exit()
        hx,hy = snake[0]; nx,ny = hx+dir[0], hy+dir[1]
        if nx<0 or nx>=COLS or ny<0 or ny>=ROWS or (nx,ny) in snake:
            return score
        snake.insert(0,(nx,ny))
        if (nx,ny)==food:
            score+=1
            if score%5==0 and speed<25: speed+=1
            food = rnd(snake)
        else:
            snake.pop()
        draw(["Use arrows or WASD to move", "ESC to quit"], snake, food, score)

while True:
    sc = game()
    s.fill((10,10,30))
    s.blit(font.render("GAME OVER",True,(230,230,230)),(W//2-40,H//2-30))
    s.blit(font.render(f"Score: {sc}  â€”  Press R to play again or ESC to quit",True,(230,230,230)),(W//2-200,H//2))
    pygame.display.flip()
    waiting = True
    while waiting:
        for e in pygame.event.get():
            if e.type==pygame.QUIT: pygame.quit(); sys.exit()
            if e.type==pygame.KEYDOWN:
                if e.key==pygame.K_r: waiting=False
                if e.key==pygame.K_ESCAPE: pygame.quit(); sys.exit()
```
